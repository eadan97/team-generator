```python
import json
from pprint import pprint

with open('data/players.json', 'r') as f:
    players = json.load(f)
pprint(players)
```

```python
for player in players:
    if "Stats" not in player:
        player["Stats"] = {"Outfield": {}, "Goalkeeper": {}}
    # player["Stats"] = {stat: value for stat, value in player.items() if stat not in ["Name", "Stats"]}
    items_for_removal = []
    for stat, value in player.items():
        match stat:
            case "Pace" | "Shooting" | "Passing" | "Dribbling" | "Defending" | "Physical":
                player["Stats"]["Outfield"][stat] = value
            case "Handling" | "Kicking" | "Reflexes" | "Speed" | "Positioning":
                player["Stats"]["Goalkeeper"][stat] = value
            case "Name" | "Stats":
                continue
        items_for_removal.append(stat)
    for stat in items_for_removal:
        del player[stat]

pprint(players)
```

```python
import yaml


with open('config.yml', 'r') as f:
        config = yaml.safe_load(f)
pprint(config)
```

```python
for player in players:
    outfield_stats = player["Stats"]["Outfield"]
    for stat, value in outfield_stats.items():
        if isinstance(value, float):
            outfield_stats[stat] = {sub_stat: value for sub_stat in config['Player Stats'][stat]}

pprint(players)
```

```python
json.dump(players, open('data/players_v2.json', 'w'), indent=4)
```
